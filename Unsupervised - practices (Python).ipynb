```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv('./data/daF3077_eng.csv', sep=';')
data.head()

## do you need to do cleaning of the data?
```

## Unsupervised learning

Unsupervised learning focuses on working without a target variable.
The goal is to find patterns in the data, such as clusters of similar observations or associations between variables.
This means there is no predefined outcome variable the algorithm tries to predict; structure is discovered from the data itself.

We examine how to cluster different welfare needs () to summarise what kind of groups emerge in the needs.

```python
variables = ["v1_1", "v1_2", "v1_3", "v1_4", "v1_5", "v1_6"]
```

```python
to_cluster = data[variables].copy()

to_cluster.describe()
```

## $k$-means

To examine this, we try to group respondents into groups based on their similarity.
For this, we use the means method, which identifies *centroids* that summarise the response patterns
of participants within each cluster, and assigns each respondent to one of the clusters.
(More of this next week.)

```python
kmeans_model = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_model.fit(to_cluster)

labels = kmeans_model.labels_
```

```python
# Distribution of respondents across clusters
pd.Series(labels).value_counts().sort_index()
```

## Analysis

We can examine the cluster centres and see their average responses.

```python
centers_df = pd.DataFrame(kmeans_model.cluster_centers_, columns=variables)
centers_df.index.name = 'cluster'
centers_df
```

## Choosing the number of clusters

To understand the number of clusters, we can use the **elbow method**.
We plot the total within-cluster sum of squares (inertia) against the number of clusters
and look for an "elbow" in the plot where the rate of decrease sharply changes.

```python
wss = [KMeans(n_clusters=k, random_state=42, n_init=10).fit(to_cluster).inertia_
       for k in range(1, 21)]

plt.figure(figsize=(7, 4))
plt.plot(range(1, 21), wss, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Within-cluster variance (inertia)")
plt.title("Elbow method")
plt.tight_layout()
plt.show()
```
