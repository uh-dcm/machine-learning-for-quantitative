```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shap

from sklearn.svm import LinearSVR

shap.initjs()   # enables interactive JS plots in Jupyter
```

```python
data = pd.read_csv("./data/daF3077_eng.csv", sep=";")

X = data[["bv1", "bv2", "bv4"]]
xvars = ["bv1", "bv2", "bv4"] 
y = data["v4_1"]
```

```python
model = LinearSVR()
model.fit(X, y)

print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
np.random.seed(10) 

to_be_examined = X.sample(n=100, random_state=10).reset_index(drop=True)
```

```python
background = shap.sample(X, 50, random_state=42)

explainer = shap.KernelExplainer(model.predict, background)
shap_values = explainer.shap_values(to_be_examined)

shap_values
```

```python
shap.summary_plot(
    shap_values,
    to_be_examined,
    plot_type="bar",
    feature_names=xvars,
    show=True
)
```

```python
shap.summary_plot(
    shap_values,
    to_be_examined,
    feature_names=xvars,
    show=True
)
```

```python
for i, var in enumerate(xvars):
    shap.dependence_plot(
        ind=i,
        shap_values=shap_values,
        features=to_be_examined,
        feature_names=xvars,
        show=True,
        title=f"Dependence plot: {var}"
    )
```
