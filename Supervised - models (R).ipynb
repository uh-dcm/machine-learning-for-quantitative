## Model tuning

In addition to various algorithms, each algorithm have parameters which impact how it operates.
They are in indicated in [the documentation](https://topepo.github.io/caret/train-models-by-tag.html).
Researchers systematically test different values out to examine what model is the best, i.e., lead to least errors.
This is called model tuning.

```r
intercept <- c(0, 1, 2, 3, 4)
```

```r
caret_model <- train(
    v4_1 ~ bv1 + bv2 + bv4,
    data = data,
    method = "lm",
    tuneGrid = data.frame(intercept = intercept)
)
```

```r
caret_model$results
```

## More parameters to tune

Above the linear model only had one paramter to tune, the intercept.
Other models can have more versatile set of parameters, leading one to establish of a grid of parameters to tune.


```r
library(nnet)
```

```r
parameters <- expand.grid(
    size = c(1, 2, 3, 4, 5),
    decay = c(0.1, 0.01, 0.001)
)

parameters
```

```r
caret_model <- train(
    v4_1 ~ bv1 + bv2 + bv4,
    data = data,
    method = "nnet",
    tuneGrid = parameters
)
```

```r
caret_model$results
```

## Choosing the model

So we can fit several models on the data, with different parameters and different values.
How to choose the best one of them?
Using the `metric` function you can try to inform what is the `finalModel`

```r
caret_model$finalModel
```

```r
caret_model <- train(
    v4_1 ~ bv1 + bv2 + bv4,
    data = data,
    method = "nnet",
    tuneGrid = parameters,
    metric = "Rsquared"
)

caret_model$finalModel
```
