## Unsupervised learning

Unsupervised learning focus on working without a target variable.
The goal is to find patterns in the data, such as clusters of similar observations or associations between variables.
This means there is no _correct_ outcome like in supervised machine learning, but rather various different outcomes.

We examine how to cluster different welfare needs (`v1_x`) to summarise what kind of groups emerge in the needs.

```r
variables <- c("v1_1", "v1_2", "v1_3", "v1_4", "v1_5", "v1_6")
```

```r
to_cluster <- data[ , variables ]
summary( to_cluster )
```

## $k$-means

To examine this, we try to group responders into groups based on their similarity.
For this, we use $k$-means method to identifies _centroids_ that summarise the response patterns of participants within each cluster (and assign each responder into one of the clusters).
(More of this next week.)

```r
clusters <- kmeans(to_cluster, centers = 3)
```

```r
table(clusters$cluster)
```

## Analysis

We can examine at the cluster centers, and see their average responses.

```r
aggregate(to_cluster, by = list(cluster = clusters$cluster), mean)
```

```r
library(factoextra)
fviz_cluster(clusters, data = to_cluster)
```

## Choosing the number of clusters

To understand the number of clusters, we can use the "elbow method"
 We can plot the total within-cluster sum of squares (WSS) against the number of clusters and look for an "elbow" in the plot where the rate of decrease sharply changes.

```r
wss <- sapply(1:20, function(k) kmeans(to_cluster, centers = k)$tot.withinss)

plot(1:20, wss, type = "b", xlab = "Number of clusters", ylab = "Within-cluster variance")
```
